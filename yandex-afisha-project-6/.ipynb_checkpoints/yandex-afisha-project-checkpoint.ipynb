```python
def df_info(df):    
    """Функция-альтернатива info(), позволяющая узнать 
    долю пропусков в столбцах датафрейма и типы данных в этих стобцах,
    а также - количество строк
    """    
    d = pd.DataFrame(df.isna().mean()).reset_index()
    f = pd.DataFrame(df.dtypes).reset_index()
    df_info = d.merge(f,on='index')
    df_info.columns = ['data', 'per_miss', 'data_type']
    return df_info, "Number of lines: {}".format(len(df))
```

### Рассчитаю `Retention Rate` с использованием когортного анализа.

* Исследуемый период составляет почти год, поэтому когорты буду брать помесячно. В случае более детального изучения какого-либо периода, например, квартала, я бы выделял этот квартал отдельно и делил его уже на недельные когорты.

```python
visits.head()
```

* Получу дату первого визита из поля `start_date` и добавлю её в новое поле для каждого пользователя.

```python
first_activity_date = visits.groupby(['uid'])['start_date'].min()
```

```python
first_activity_date.head(2)
```

```python
first_activity_date.name = 'first_start_date'
```

```python
visits = visits.join(first_activity_date,on='uid')
```

* Получу месяц первого визита (месячная когорта)

```python
visits['first_start_month'] = visits['first_start_date'].astype('datetime64[M]')
```

* Получу `lifetime` месячных когорт.

```python
visits['cohort_lifetime'] = (
    (visits['start_month'] - visits['first_start_month']) 
    / np.timedelta64(1,'M')
    )
```

```python
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
```

* Сгруппирую данные по месяцу первой сессии и `cohort_lifetime`, чтобы получить количество уникальных пользователей месячных когорт на старте и в последующие отрезки `cohort_lifetime`

```python
cohorts = (
    visits.groupby(['first_start_month','cohort_lifetime']).
    agg({'uid':'nunique'}).reset_index()
)
```

```python
cohorts.head()
```

* Получу число уникальных пользователей в когорте на момент её формирования и добавлю в таблицу `cohorts`

```python
initial_users_count = (
    cohorts[cohorts['cohort_lifetime'] == 0][['first_start_month','uid']]
)
```

```python
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
```

```python
cohorts = cohorts.merge(initial_users_count, on='first_start_month')
```

* Получу отношение числа пользователей по каждому месяцу к их первоначальному количеству в когорте

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
```

```python
cohorts.head()
```

* Построю сводную таблицу и тепловую карту `Retention Rate`

```python
retention_pivot = cohorts.pivot_table(
    index='first_start_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum'
)
```

```python
retention_pivot.fillna('')
```

```python
plt.figure(figsize=(14,9))
plt.title('Удержание пользователей в когортах')
sns.heatmap(
    retention_pivot,
    annot=True,
    fmt='.2%',
    linewidths=0.5,
    linecolor='white',
    cmap='gist_heat_r'
)
plt.ylabel('Месячная когорта')
plt.show()
```

* На втором месце жизни когорт можно заметить большую разницу в цвете тёплых летних и первых осенних месяцев относительно более холодных месцев (ноябрь, зима, март).
* Вероятно, это связано с большей аткивнойстью пользователей в тёплое время года: и ведь так и есть - милее погода, больше прогулок и мероприятий посещаем.


---


* Получу среднее значение `Retention Rate` на второй месяц жизни когорт.

```python
(print(
    'Средний процент удержания пользователей на второй месяц: {:.2%}'.
    format(retention_pivot[1].mean()))
)
```

### Расчитаю среднее количество покупок на одного покупателя за выбранный период (6 месяцев).

* Выделю когорты пользователей на основе времени первой покупки
* Посмотрю как изменяется число покупок в когортах на протяжении времени.

* Получу месяц покупки из поля `buy_ts`

```python
orders['buy_month'] = orders['buy_ts'].astype('datetime64[M]')
```

```python
orders.head(2)
```

* Получу количество уникальных пользователей и число покупок в первые пол года исследуемого периода.
* Разделю второе на первое и получу среднее количество покупок на одного пользователя за пол года.

```python
orders_by_users = (
    orders[(orders['buy_month'] >= '2017-06-01') 
    & (orders['buy_month'] < '2017-12-01')]
    .agg({'uid':['nunique','count']})
    .reset_index()
    )
```

```python
orders_by_users
```

```python
orders_by_users['uid'][1] / orders_by_users['uid'][0]
```

* `1.3` покупки приходится в среднем на пользователя в первое полугодие исследуемого периода (с 1 июня 2017 по 1 декабря 2017)

```python
orders.head()
```

* Сделаю выборку данных о покупках за первые пол года исследуемого периода и посчитаю, сколько покупок было у каждого покупателя в месяце.

```python
orders_per_user_half_year = (
    orders[(orders['buy_month'] >= '2017-06-01') 
    & (orders['buy_month'] < '2017-12-01')].
    groupby(['buy_month','uid'])['buy_ts'].
    count().reset_index()
)
```

* Посчитаю, сколько уникальных покупателей было в месяц и сколько в каждом месяце они совершили покупок

```python
avg_purchase_per_user = (
    orders_per_user_half_year.
    groupby('buy_month').
    agg({'uid':'count', 'buy_ts':'sum'}).
    reset_index()
)
```

```python
avg_purchase_per_user
```

* Получу среднее количество покупок на одного покупателя в месяц.

```python
avg_purchase_per_user['ppu'] = (
    avg_purchase_per_user['buy_ts'] / 
    avg_purchase_per_user['uid']
)
```

### Рассчитаю средний LTV на покупателя по когортам за 6 месяцев

```python
orders.head()
```

* Поле с месяцем любой покупки у меня уже есть. Получу поле с месяцем первой покупки.

```python
first_orders = orders.groupby('uid').agg({'buy_month':'min'}).reset_index()
```

```python
first_orders.head()
```

```python
first_orders.columns = ['uid', 'first_buy_month']
```

* Посчитаю количество новых покупателей за каждый месяц `n_buyers`

```python
cohort_sizes = (
    first_orders.groupby('first_buy_month').
    agg({'uid':'nunique'}).
    reset_index()
)
```

```python
cohort_sizes
```

```python
cohort_sizes.columns = ['first_buy_month', 'n_buyers']
```

* Создам таблицу  заказов, содержащую месяц первой покупки

```python
orders_with_first_buy_month = orders.merge(first_orders, on='uid')
```

```python
orders_with_first_buy_month.head()
```

* Сгруппирую данные о выручке по месячным покупательским когортам

```python
orders_cohorts = (
    orders_with_first_buy_month.
    groupby(['first_buy_month','buy_month']).
    agg({'revenue':'sum'}).reset_index()
)
```

```python
orders_cohorts.head()
```

* Добавлю в таблицу `orders_cohorts` данные о численности когорт

```python
orders_cohorts = orders_cohorts.merge(cohort_sizes,on='first_buy_month')
```

```python
orders_cohorts.head()
```

* LTV считается из из валовой прибыли, которая связана с выручкой через маржинальность. В данном случае маржинальность равна максимальному значению 100%. Это означает, что валовая прибыль в этом случае равна выручке.

```python
orders_cohorts['ltv'] = orders_cohorts['revenue'] / orders_cohorts['n_buyers']
```

```python
orders_cohorts.head()
```

* Добавлю поле `cohort_lifetime`

```python
orders_cohorts['cohort_lifetime'] = (
    (orders_cohorts['buy_month'] - 
     orders_cohorts['first_buy_month']) / np.timedelta64(1, 'M')
)
```

```python
orders_cohorts['cohort_lifetime'] = (
    orders_cohorts['cohort_lifetime'].
    round().astype('int')
)
```

```python
orders_cohorts.head()
```

* Отфильтрую когорты прожившие не менее 6 месяцев (созданные не позднее ноября 2017)

```python
orders_cohorts = orders_cohorts[orders_cohorts['first_buy_month'] <= '2017-12-01']
```

* Создам сводную таблицу с накопительным `LTV`

```python
output = orders_cohorts.pivot_table(
    index='first_buy_month',
    columns='cohort_lifetime',
    values='ltv',
    aggfunc='mean'
)
```

```python
output = output.cumsum(axis=1).round(2)
```

```python
output.fillna('')
```

* Выведу эти данные на тепловой карте

```python
plt.figure(figsize=(14,9))
plt.title('Средний ltv пользователей в когортах')
sns.heatmap(
    output,
    annot=True,
    fmt='.2f',
    linewidths=0.5,
    linecolor='white',
    cmap='gist_heat_r'
)
plt.ylabel('Месячная когорта')
plt.show()
```

* Когорта покупателей сентября 2017 обладает наиболее высоким средним `ltv`. Вероятно, стоит обратить внимание на характеристики аудитории, которая тогда привлекалась через рекламу.
* Следом за ней по приоритету должна идти когорта июня 2017: у них тоже весьма неплохой `ltv`
* **Но это предварительные выводы: надо ещё рассчитать `ROMI`, для которого когорты должны формироваться не по первой покупке, а по первому визиту привлечённого покупателя.**
* А пока получу средний `ltv` на 6-ой месяц жизни покупательских когорт.

```python
(
    print(
        'Средний ltv пользователя на 6-ой месяц жизни покупательской когорты: {:.2f} у.е.'.
        format(output[5].
               mean()))
)
```

```python
buyers_daily['first_order_dt'] = pd.to_datetime(buyers_daily['first_order_dt'])
```

- Объединю таблицы

```python
for_cac_example = (
    pd.merge(
        buyers_daily, 
        costs, 
        left_on=['source_id', 'first_order_dt'], 
        right_on=['source_id', 'dt'])
)
```

```python
for_cac_example['cac'] = for_cac_example['costs']/for_cac_example['n_buyers']
```

- Построю график

```python
(
    for_cac_example.
    pivot_table(
        index=['dt_month'], 
        columns='source_id', 
        values='cac', 
        aggfunc='mean'
    ).plot(
        figsize=(16,9), 
        linewidth=2, 
        title='Стоимость привлечения одного покупателя из каждого источника')
)
plt.show()
```

### Рассчитайтаю ROMI по когортам в разрезе источников. Сравню окупаемость за одинаковые периоды жизни когорт.

* Для начала мне потребуется `LTV` когорт с разбивкой по источникам:
    * Для его расчёта использую полученную ранее таблицу `orders_sources`, которая содержит:
        * дату, месяц первой покупки и источник первой покупки

```python
orders_sources
```

* Добавлю столбцы c месяцем первой покупки и первым источником покупателя к таблице заказов `orders`

```python
orders_with_source = orders.merge(orders_sources,on='uid')
```

```python
orders_with_source
```

* Для удобства пересохраню таблицу без лишних на данном этапе полей 

```python
orders_with_source = (
    orders_with_source[['uid', 'source_id', 'buy_month', 
                        'first_order_month', 'revenue']].
    copy()
)
```

```python
orders_with_source
```

* Посчитаю количество новых покупателей по каждому источнику, месяцу первой покупки(когорте), месяцу каждой покупки

```python
cohorts_size = (
    orders_with_source.
    groupby(['source_id','first_order_month']).
    agg({'uid':'nunique'}).
    reset_index()
)
```

```python
cohorts_size.rename(columns={'uid':'n_buyers'}, inplace=True)
```

```python
cohorts_size
```

```python
print(cohorts_size['n_buyers'].sum())
print(orders['uid'].nunique())
```

* Покупатели на месте
* Теперь сформирую когорты по месяцу первого визита, месяцу каждой покупки и первому источнику и просуммирую `revenue` по каждой группе

```python
cohorts = (
    orders_with_source.
    groupby(['source_id','first_order_month','buy_month']).
    agg({'revenue':'sum'}).
    reset_index()
)
```

```python
cohorts
```

```python
print(orders['revenue'].sum())
print(cohorts['revenue'].sum())
```

* Выручка на месте
* Теперь каждой когорте добавлю данные о том, сколько в ней покупателей

```python
report = cohorts.merge(
    cohorts_size,
    how='left',
    on=['first_order_month','source_id']
)
```

```python
report
```

```python
print(orders['revenue'].sum())
print(report['revenue'].sum())
```

* Теперь рассчитаю `LTV` по каждой когорте
* Помню, что валовая прибыль, необходимая для рассчёта `LTV`, в случае 100%-й маржинальности равна самой выручке

```python
report['ltv'] = report['revenue'] / report['n_buyers']
```

```python
report.head()
```

* Добавлю `cohort_age`

```python
report['cohort_age'] = (
    (report['buy_month'] - 
    report['first_order_month']) / np.timedelta64(1,'M')
)
```

```python
report['cohort_age'] = report['cohort_age'].round().astype('int')
```

* Добавлю данные, нужные для расчёта `CAC` и `ROMI` к таблице `report`
* Это будут данные из `costs`, сгруппированные по месяцам и источникам

```python
group_costs = (
    costs.groupby(['dt_month','source_id']).
    agg({'costs':'sum'}).
    reset_index()
)
```

```python
group_costs
```

```python
report = (
    report.merge(
        group_costs, 
        how='left', 
        left_on=['source_id','first_order_month'], 
        right_on=['source_id','dt_month'])
)
```

```python
report
```

```python
report['cac'] = report['costs'] / report['n_buyers']
```

```python
report['romi'] = report['ltv'] / report['cac']
```

```python
report[report['cac'].isna()]
```

* Есть отсутствующее значение в полях, которые рассчитывались с участием данных таблицы `costs`
* Это та самая покупка от пользователя, пришедшего с бесплатного 7-го источника. И покупка сделанная в первые минуты июня 2018
* Они в построении графиков не помешают.

```python
report
```

* Получил таблицу-заготовку, из которой уже можно строить сводные по месячным когортам и источникам:
* Я подумал и решил, что `ЧТО БЫ Я ТАМ НИ РЕШАЛ РАНЕЕ` надо сделать отдельные тепловые карты по каждому источнику)):
   
     * Поэтому выведу сперва график накопительного `ROMI` по времени жизни когорт
     * Следом выведу такие же графики по каждому источнику отдельно

```python
output_all_source = report.pivot_table(
    index='first_order_month',
    columns='cohort_age',
    values='romi',
    aggfunc='mean'
)
```

```python
output_all_source
```

* Посчитаю накопительный `ROMI` по месячным когортам в течение их дальнейшей "жизни"

```python
output_all_source_cs = output_all_source.cumsum(axis=1).round(2)
```

```python
output_all_source_cs
```

```python
plt.figure(figsize=(14,9))
plt.title('Накопительный ROMI по месячным когортам в течение их жизни')
sns.heatmap(
    output_all_source_cs,
    annot=True,
    fmt='.2f',
    linewidths=0.5,
    linecolor='white',
    cmap='gist_heat_r'
)
plt.ylabel('Когорты')
plt.xlabel('Месяц жизни когорты')
plt.show()
```

* Первая когорта, вторая и четвёртая когорты вышли на "безубыточность" по итогу 4-го месяца и продолжили рост:
    * На них маркетологам следует сделать упор, если возможно - сформировать `Lool-a-like` аудиории и потестировать их.
    
* Посмотрю, как менялся `ROMI` когорт в рамках каждого источника отдельно

```python
source_list = [1,2,3,4,5,9,10]
```

```python
for i in source_list:
    output_i = report[report['source_id'] == i].pivot_table(
    index='first_order_month',
    columns='cohort_age',
    values='romi',
    aggfunc='mean'
    )
    output_i = output_i.cumsum(axis=1).round(2)
    plt.figure(figsize=(14,9))
    plt.title('Накопительный ROMI по когортам из источника {} в течение их жизни'.format(i))
    sns.heatmap(
    output_i,
    annot=True,
    fmt='.2f',
    linewidths=0.5,
    linecolor='white',
    cmap='mako_r'
    )
    plt.ylabel('Когорты')
    plt.xlabel('Месяц жизни когорты')
    plt.show()
```
