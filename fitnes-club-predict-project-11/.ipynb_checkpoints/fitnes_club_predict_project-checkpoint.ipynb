```python
def df_info(df):    
    """Функция-альтернатива info(), позволяющая узнать 
    долю пропусков в столбцах датафрейма и типы данных в этих стобцах,
    а также - количество строк и столбцов.
    """    
    d = pd.DataFrame(df.isna().mean()).reset_index()
    f = pd.DataFrame(df.dtypes).reset_index()
    df_info = d.merge(f,on='index')
    df_info.columns = ['data', 'per_miss', 'data_type']
    return df_info, "Rows x Columns: {} x {}".format(df.shape[0], df.shape[1])
```

## Построю модель прогнозирования оттока клиентов

---

- Разделю данные на признаки и целевую переменную.

```python
X = df.drop('churn', axis=1)
```

```python
y = df['churn']
```

- Разобью данные на обучающую и валидационную выборки.

```python
X_train, X_val, y_train, y_val = (
    train_test_split(X, y, test_size=0.3, random_state=0)
)
```

- Обучу модель на train-выборке логистической регрессией.
  - Сперва с признаками с сильной корреляцией.
  - Затем отбросив один из каждой пары.

```python
lr_model = LogisticRegression(random_state=0)
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_predictions = lr_model.predict(X_val)
```

```python
lr_probabilities = lr_model.predict_proba(X_val)[:,1]
```

- Добавлю функцию, которая будет считать и выводить все необходимые метрики.

```python
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

- Теперь пересохраню обучающую и валидационную выборки без признаков с сильной корреляцией.
- Переобучу модель на тренировочной выборке без скоррелированных признаков.
- Сделаю прогноз на валидационной выборке без скоррелированных признаков.
- Проверю как поменялись метрики классификации, относительно прогноза со всеми признаками.

```python
X_train_no_corr = X_train.drop(
    [
        'avg_class_frequency_total', 
        'contract_period'
    ], axis=1
)
```

```python
X_val_no_corr = X_val.drop(
    [
        'avg_class_frequency_total', 
        'contract_period'
    ], axis=1
)
```

```python
lr_model.fit(X_train_no_corr, y_train)
```

```python
lr_no_corr_predictions = lr_model.predict(X_val_no_corr)
```

```python
lr_no_corr_probabilities = lr_model.predict_proba(X_val_no_corr)[:,1]
```

- Обучу модель на train-выборке "Случайным Лесом" .
  - Сперва с признаками с сильной корреляцией.
  - Затем отбросив один из каждой пары.

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
```

```python
rf_model.fit(X_train, y_train)
```

```python
rf_predictions = rf_model.predict(X_val)
```

```python
rf_probabilities = rf_model.predict_proba(X_val)[:,1]
```

- Теперь переобучу модель без сильно скоррелированных признаков и сделаю на ней прогноз

```python
rf_model.fit(X_train_no_corr, y_train)
```

```python
rf_no_corr_predictions = rf_model.predict(X_val_no_corr)
```

```python
rf_no_corr_probabilities = rf_model.predict_proba(X_val_no_corr)[:,1]
```

- Выведу метрики для всех моделей случайного леса и логистической регрессии, обученных на всех признаках и без сильно скоррелированных признаков.

```python
print_all_metrics(
    y_val,
    lr_predictions,
    lr_probabilities,
    title='Метрики для модели логистической регрессии:',
)
print_all_metrics(
    y_val,
    rf_predictions,
    rf_probabilities,
    title='Метрики для модели случайного леса:',
)
print_all_metrics(
    y_val,
    lr_no_corr_predictions,
    lr_no_corr_probabilities,
    title='Метрики для модели логистической регрессии без сильно скоррелированных признаков:',
)
print_all_metrics(
    y_val,
    rf_no_corr_predictions,
    rf_no_corr_probabilities,
    title='Метрики для модели случайного леса без сильно скоррелированных признаков:',
)
```

### Итоги сравнения моделей логистической регрессии и случайного леса: 
Со всеми признаками, а также с исключением сильно скоррелированных(один из пары).

- Доля правильных ответов или `Accuracy` получила наибольшее и одинаковое значение как для логистической регрессии, так и для случайного леса. Но важно отметить, что произошло это с использованием всех признаков без исключения сильно скоррелированных. 

(Тут я ещё раз внимательно прочёл расздел про регуляризацию и понял, что он относится к моделям линейной регресии, а не логистической регрессии или случайному лесу. То есть отбрасывать признаки с высокой корреляций было не обязательно. Но проверить было интересно. Ну или ревьюэр мне скажет, где я промахнулся в рассуждениях) 

- Верная доля прогнозов относительно класса `"1"` или `Precision` имеет наибольшее значение у модели логистической регрессии и составляет `0,86`. Но в нашем случае мы не боимся ложного срабатывания и отнесения человека к сегменту оттока. Это не выдача кредита неблагонадёжному заёмщику. Самое страшное, что может случиться: мы ещё один раз обратимся к нашему клиенту, думая, что он хочет с нами расстаться, хотя это может быть не так. Покажем ему рекламное сообщение о выгодной акции и предложим маленький бонус - отлично. Главное, не делать это каждый день и знать меру (её тоже лучше посчитать и спрогнозировать).

- Поэтому более интересной метрикой в нашем случае выступает `Recall`: сколько реальных объектов класса `"1"` мы смогли обнаружить. Полезна при диагностировании болезней, к коим можно отнести аппатию, лень и отвращение к регулярным физическим нагрузкам. Иными словами, лучше мы сочтём случайно "беглецом" того, кто не планирует бежать, и пораблотаем над его лояльностью немного дополнительно.

---

- А определять, кто скоро сбежит, а кто нет буду с помощью модели-победителя `RandomForestClassifier`, потому что она показала большую величину метрики, приоритетной в нашей задаче `Recall: 0.84` при равных у обеих моделей `Accuracy` и `ROC_AUC`.


 

## Кластеризация клиентов

- Стандартизирую данные признаков


```python
sc = StandardScaler()
```

```python
X_sc = sc.fit_transform(X)
```

- Построю матрицу расстояний перед построением дендрограммы

```python
linked = linkage(X_sc, method = 'ward')
```

- Строим дендрограмму 

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.show()
```

- Предполагаемое оптимальное число кластеров 4 (количество цветов на графике).

---

- **Обучу модель кластеризации с помощью алгоритма K-Means и спрогнозирую кластеры клиентов.** 
  - за значение параметра `n_clusters` примем 5

```python
km = KMeans(n_clusters = 5, random_state=0)
```

```python
labels = km.fit_predict(X_sc)
```

```python
df['cluster_km'] = labels
```

- Посмотрю на средние значения признаков для кластеров

```python
clustered = df.groupby(['cluster_km'], as_index=False).mean()
```

```python
clustered
```

 - В таблице средних значений признаков по кластерам можно заметить следующее:
  - с учётом заданного числа кластеров 5 достаточно хорошо просматривается связь между средним числом тренировок в неделю и долей оттока: больше число тренировок - меньше доля оттока.
  - аналогичная взаимосвязь между признаком `lifetime` и долей оттока: чем больше мы дружим с клиентом, тем чаще он ходит в среднем в зал, тем менее вероятно покинет нас.
  - уровень расходов также отражается на потенциальном желании покинуть мир физкультуры. Те, кто уже давно осознал приобретаемую ценность, тратят на сопутствующие товары и услуги в среднем больше.

### Построю диаграммы разброса для пересечения некоторых небинарных признаков с подсветкой кластеров.

  - Для этого задам функцию `show_clusters_on_plot`

```python
def show_clusters_on_plot(df, x_name, y_name, cluster_name):
    plt.figure(figsize=(10, 10))
    sns.scatterplot(
        df[x_name], 
        df[y_name], 
        hue=df[cluster_name], 
        palette='Paired'
    )
    plt.title('{} vs {}'.format(x_name, y_name))
    plt.show()
```

- Посмотрим на диаграмму разброса кластеров на пересечении признаков:
  - средняя частота посещений в неделю за все время
  - месяцев до конца абонемента

```python
show_clusters_on_plot(
    df,
    'avg_class_frequency_total',
    'month_to_end_contract',
    'cluster_km'
)
```

- **О чём говорит график**
  - Самый надёжный, 0-й кластер разместился в наиболее логичной зоне: там, где ещё остались предоплачены от 7 до 12 месяцев.
  - Следующий за ним размесмтился в области частых тренировок. Люди увлечённые.
  - Самый ненадёжный расположился ближе к левому краю графика, где визиты редки, а до конца абонемента от 6 до 1 месяца (большинство с лёгкостью прощаются с последним оплаченным месяцем) 

- Посмотрим на диаграмму разброса кластеров на пересечении признаков:
  - возраст
  - средняя частота посещений в неделю за все время

```python
show_clusters_on_plot(
    df,
    'age',
    'avg_class_frequency_total',
    'cluster_km'
)
```

- **О чём говорит график**
  - Самый надёжный, 0-й тренирующейся кластер на этот раз равномерно распределён по возрастам и частоте тренинга, а вот 4-й, следующий за ним по надёжности, заметно смещён в сторону публики постарше, тренирующейся чаще.
  - Самый ненадёжный 3-й смещён в сторону беспечной молодости и лени (редкие тренировки).

- Посмотрим на диаграмму разброса кластеров на пересечении признаков:
  - траты на сопутствующие товары и услуги
  - средняя частота посещений в неделю за все время

```python
show_clusters_on_plot(
    df,
    'avg_additional_charges_total',
    'avg_class_frequency_total',
    'cluster_km'
)
```

- **О чём говорит график**
  - Самый надёжный, 0-й тренирующейся кластер на этот раз тоже неявно распределён по тратам и частоте тренинга, а вот 4-й, следующий за ним по надёжности, больше смещён в сторону трат побольше, тренинга чаще.
  - Самый ненадёжный 3-й смещён в сторону редких тренировок и минимальных сопутствующих затрат.

## Выводы исследования и базовые рекомендации по работе с клиентами

1. Уже проанализировав средние значения признаков выборки, разделённой по критерию оттока можно сделать предварительные выводы путём простой логики. Например:
  - Среди оставшихся пользователей больше доля тех, кто проживает или работает в пределах района фитнес-центра.
  - Среди оставшихся пользователей больше доля тех, кто тренируется по партнёрским программам (скидки на абонемент).
  - Среди оставшихся пользователей больше доля тех, кто пришёл по акции "приведи друга". Очевидно, что с друзьями тренироваться веселее, больше мотивация продолжать.
  - Среди оставшихся пользователей больше доля тех, чей действующий абонемент 6 месяцев. В группе оттока напротив - больше доля коротких абонементов на месяц.

2. Определять, кто скоро сбежит, а кто нет неплохо получается с помощью модели `RandomForestClassifier`, потому что она показала большую величину метрики, приоритетной в нашей задаче `Recall: 0.84` при равных у обеих моделей конкурсантов `Accuracy` и `ROC_AUC`. Приоритет при прочих равных отдан метрике `Recall` неспроста. Она нацелена на минимизацию противоположных рисков. Иными словами:
  - Лучше мы провзамодействуем лишний раз с тем, кто уходить не собирался, чем будем бояться потревожить клиентов выгодным предложением и упустим тех, кто склонен к уходу.
  - Важно отметить, что влиятельная для оценки качества классификатора вероятностная метрика `ROC_AUC` для обеих моделей была на хорошем уровне `0.97`.

3. Дендрограмма построенная по матрице расстояний признаков показала 4 кластера клиентов.
  - Такое же значение я предположил, основываясь на значениях долей оттока, полученных группировкой признака оттока по кластерам. Но в академических целях алгоритму K-Means было задано число кластеров 5.

4. На диаграммах рассеяния лучше всего прослеживается взаимосвязь признаков и распределение кластеров при пересечении небинарных признаков. Иначе всё просто сольётся в 2 разноцветные полосы. На основе этих зависимостей можно предложить следующие маркетинговые действия в отношении самого подверженного оттоку 3-го кластера:
  - если период до окончания абонемента 6 и менее месяцев, можно предлагать за каждое 3-е занятие в неделю ощутимую скидку на услуги массажиста (баня и сауна для восстановления должны уже быть включены). Таким образом мотивируем клиента чаще заходить, тратить дополнительно деньги(потому что существенная скидка) на массажный кабинет. За счёт этого он будет лучше восстанавливаться, больше тренироваться и получать более ощутимый результат от тренировок. А значит, желание бросить будет сокращаться.
  - для клиентов, у которых признак `lifetime` менее года следует чаще устраивать групповые бонусные тренировки, потому что участие в групповых тренировках добавляет соревновательный момент и азарт в тренинг. Что положительно сказывается желании остаться и ходить чаще. Собственно, в этом феномен кроссфита и его стремительного распространения. Помимо простоты инвентаря и функциональности упражнений, он в первую очередь командный: все всегда тренируются в группах, неявно но соревнуются даже в процессе тренировки.
  - На вышеописанные маркетинговые действия можно применить повышающий коэффициент для аудитории более молодой, до 30-лет. Многие в этом возрасте ещё весьма беспечно относятся к регулярным тренировкам и не осознают их полной ценности. Поэтому мотивировать их стоит сильнее, чем публику из диапазона 30-40 лет.
  - После применения этих мер на самый подверженный оттоку кластер стоит собрать данные (хотя бы за 3 месяца) проанализировать (не забываем про статистическую погрешность) как изменился показатель оттока в этом кластере. Если меры принесут ощутимый положительный результат, можно пробовать масштабировать их на соседние, чуть менее подверженные оттоку кластеры, естественно, учитывая экономическую целесообразность.
